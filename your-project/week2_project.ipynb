# FUNTION HOURS' ANALYSIS

```python
#this is giving us the mean of each hour in the whole month
month_hour = []
#print('This loop will show the mean value of each hour during the month')

def hours_through_month(neighbourhood_and_contaminant_table):
    month_hour = []
    table_list = neighbourhood_and_contaminant_table
    for hour in range(1,25):
        # print(f'Hour number {hour}.') # test print
        if hour < 10:
            print(table_list.loc[:,'H0'+str(hour)].mean()) # test print
            month_hour.append(table_list.loc[:,'H0'+str(hour)].mean())
        else: 
            print(table_list.loc[:,'H'+str(hour)].mean()) # test print
            month_hour.append(table_list.loc[:,'H'+str(hour)].mean())
    return pd.Series(month_hour)
        
#estacio4_cont8 = hours_through_month(tables_list[1])       
#estacio4_cont8
#max(month_hour)
#min(month_hour)

```

# FUNCTION DAYS' ANALYSIS

```python

def days_in_month(neighbourhood_and_contaminant_table):

    #we will find sometimes that a list will have less elements than days. 
    #This will be because it will have days full of NaN numbers.

    #meanperhours during the month
    month_day = []

    #print(tables_list[1].DIA)
    #print('This loop will show the mean value of each day during the month')
    table = neighbourhood_and_contaminant_table
    table = table.reset_index(drop=True)
    #print(table) #test print

    #testing prints
    #print(table.loc[0,'H05'])
    #print('______')

    # print (len(table.DIA)) # print test

    for day in range(0, len(table.DIA)):
        #print(day) #print test
        table = table.fillna(0)
        month_day = []
        total_day = 0
        hours_counter = 0
        for hour in range(1,25):
            #print(f'Hour number {hour}.')  #print test
            #print('the day is', day)  #print test
            if hour < 10:
                if table.loc[day,'H0'+str(hour)] != 0:
                    total_day += table.loc[day,'H0'+str(hour)]
                    hours_counter += 1
                    #print(total_day) #print test
                    #print (hours_counter)
            else:
                if table.loc[day,'H'+str(hour)] != 0:
                    total_day  += table.loc[day,'H'+str(hour)]
                    hours_counter += 1
                    #print(total_day) #print test
                    #print (hours_counter)
            if hours_counter == 0:
                hours_counter = 1

            month_day.append(total_day/hours_counter)
    
    
    #print(len(month_day) )       
    #print(month_day)
    #variable_intermedia = pd.Series(month_day)
    #variable_intermedia.to_csv('contaminante.csv')
    return month_day
    
```
